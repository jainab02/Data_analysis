# Iris data set analysis

```python
#importing the libraries of python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Exploration

reading the dataset to analyze the data from it

```python
df = pd.read_csv('Iris.csv')
df
```

printing the first 5 rows of the dataset

```python
#printing the first 5 rows of the dataset
df.head()
```

Some basic information about the data on which we are working

```python
#shape gives total no of columns and rows
df.shape
```

```python
#it gives us the name of all the columns
df.columns
```

```python
#info gives us the summary of dataset 
#including column name null or not null and datatype 
df.info()
```

```python
#describe gives us basic statistical values 
#like count,mean,standard deviation,minimum etc
df.describe()
```

Cleaning of the data

```python
#checking for missing values in the dataset
#isnull() method is used for this purpose
df.isnull().sum()
```

```python
#Removing for duplicates
#drop_duplicates() method is used for this purpose
d = df.drop_duplicates(subset ="Species",)
d
```

```python
#value_count() is used for checking if our dataset is balanced or not 
df.value_counts("Species")
```

```python
#skewness 
df.skew()
```

```python
#kurtosis
df.kurtosis()
```

```python
#correlation
df.corr()
```

# Data visualization

```python
df.plot()
```

```python
#ploting the graph of species and their
#here we use seaborn as it has high statistical graphics than matplotlib
sns.countplot(x='Species', data=df )
plt.title('species count')
plt.show()
```

```python
sns.pairplot(df,"Species",size=2)
```

```python
#ploting line graph for comparing sepal width and sepal length
sns.lineplot(df['SepalLengthCm'],df['SepalWidthCm'])
plt.title("comparision between sepal width and length")
```

```python
#ploting line graph for comparing petal width and petal length
sns.lineplot(df['PetalLengthCm'],df['PetalWidthCm'])
plt.title("comparision between petal width and petal length")
```

```python
#histograph for first 10 entries of sepal length
plt.title("histograph of sepal length")
plt.hist(df["SepalLengthCm"],bins=10)
```

```python
#histograph for first 10 entries of sepal width
plt.title("histograph of sepal width")
plt.hist(df["SepalWidthCm"],bins=10)
```

```python
#histograph for first 10 entries of petal length
plt.title("histograph of petal length")
plt.hist(df["PetalLengthCm"],bins=10)
```

```python
#histograph for first 10 entries of petal width
plt.title("histograph of petal width")
plt.hist(df["PetalWidthCm"],bins=10)
```
